```python
from Bio import Entrez
from Bio.Entrez import efetch, read
Entrez.email = "rio101@email.unc.edu"
```

```python
def fetch_abstract(pmid):
    handle = efetch(db='pubmed', id=pmid, retmode='xml')
    xml_data = Entrez.read(handle)
    try:
        abstract = xml_data['PubmedArticle'][0]['MedlineCitation']['Article']['Abstract']
        return abstract
    except IndexError:
        return None
    except KeyError:
        return None
```

```python
no_grant = []
def fetch_grant(pmid):
    grant_id = []
    handle = efetch(db='pubmed', id=pmid, retmode='xml')
    xml_data = Entrez.read(handle)
    try:
        for i in range (0, len(xml_data['PubmedArticle'][0]['MedlineCitation']['Article']['GrantList'])):
            grant = xml_data['PubmedArticle'][0]['MedlineCitation']['Article']['GrantList'][i]['GrantID']
            agency = xml_data['PubmedArticle'][0]['MedlineCitation']['Article']['GrantList'][i]['Agency']
            grant_id.append(agency + ', ' + grant)
        return grant_id
    except IndexError:
        return None
    except KeyError:
        if pmid not in no_grant:
            no_grant.append(pmid)
        return no_grant
```

```python
fetch_grant(34610516)
```

```python
 def fetch_keyw(pmid):
    handle = efetch(db='pubmed', id=pmid, retmode='xml')
    xml_data = Entrez.read(handle)
    try:
        keyw = xml_data['PubmedArticle'][0]['MedlineCitation']['KeywordList'][0][0]
        return keyw
    except IndexError:
        return None
```

```python
fetch_keyw(33531212)
```

```python
def fetch_pub(pmid):
    handle = efetch(db='pubmed', id=pmid, retmode='xml')
    xml_data = Entrez.read(handle)
    try:
        pub = xml_data['PubmedArticle'][0]['MedlineCitation']['Article']['Journal']['JournalIssue']['PubDate']['Year']
        return pub
    except IndexError:
        return None
    except KeyError:
        return '0000'
```

```python
fetch_pub(33531212)
```

```python
def fetch_name(pmid):
    handle = efetch(db='pubmed', id=pmid, retmode='xml')
    xml_data = Entrez.read(handle)
    try:
        name = xml_data['PubmedArticle'][0]['MedlineCitation']['Article']['ArticleTitle']
        return name
    except IndexError:
        return None
```

```python
fetch_name(33531212)
```
